```python
model.fit(X_train, y_train)
```

```python
#5. Find accuracy, confusion matrix and classification report
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
df.columns = ["Class", "Left_Weight", "Left_Distance", "Right_Weight", "Right_Distance"]
```

```python
print(df.columns)
```

```python
X = df[["Left_Weight", "Left_Distance", "Right_Weight", "Right_Distance"]]
```

```python
y = df["Class"]
```

```python
print("Input Features:")
```

```python
print("\nOutput Target:")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(
    random_state=42
)
```

```python
dt_model.fit(X_train, y_train)
```

```python
#5. Find accuracy, confusion matrix and classification report 6.
#Perform the prediction of single value entry.
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
y_pred = dt_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
print("\nConfusion Matrix:")
```

```python
print(cm)
```

```python
print("\nClassification Report:")
```

```python
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(15, 10))

plot_tree(
    dt_model,
    feature_names=X.columns,
    class_names=dt_model.classes_,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree Visualization")
plt.savefig("decision_tree.png", dpi=300, bbox_inches="tight")
plt.show()

```
