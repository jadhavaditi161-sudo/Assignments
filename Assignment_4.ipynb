```python
class Student:

    school = "ABC College"

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks


    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)
        print("School:", Student.school)

    @classmethod
    def change_school(cls, new_school):
        cls.school = new_school

    @staticmethod
    def grade(marks):
        if marks >= 75:
            return "A"
        elif marks >= 50:
            return "B"
        else:
            return "C"
```

```python
s1 = Student("Aditi", 85)
```

```python
s2 = Student("Rahul", 60)
```

```python
print("Before changing school:")
```

```python
s1.display()
```

```python
Student.change_school("XYZ College")
```

```python
print("\nAfter changing school:")
```

```python
s2.display()
```

```python
print("\nGrades:")
```

```python
print("Aditi Grade:", Student.grade(85))
```

```python
print("Rahul Grade:", Student.grade(60))
```

```python
class Student:

    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks    # Private variable


    def get_marks(self):
        return self.__marks


    def set_marks(self, marks):
        if marks >= 0 and marks <= 100:
            self.__marks = marks
        else:
            print("Invalid marks! Enter marks between 0 and 100.")


    def display(self):
        print("Name:", self.name)
        print("Marks:", self.__marks)
```

```python
#14.	Write a Python program to create an abstract class, implement abstract methods, and instantiate child classes.
```

```python
from abc import ABC, abstractmethod
```

```python
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass
```

```python
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width
```

```python
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius
```

```python
rect = Rectangle(10, 5)
```

```python
circle = Circle(7)
```

```python
print("Area of Rectangle:", rect.area())
```

```python
print("Area of Circle:", circle.area())
```

```python
#16.	Write a Python program to overload arithmetic operators, compare objects, and display the results.
```

```python
class Number:
    def __init__(self, value):
        self.value = value


    def __add__(self, other):
        return Number(self.value + other.value)


    def __sub__(self, other):
        return Number(self.value - other.value)


    def __eq__(self, other):
        return self.value == other.value


    def __gt__(self, other):
        return self.value > other.value


    def __str__(self):
        return str(self.value)

```

```python
n1 = Number(20)
```

```python
n2 = Number(10)
```

```python
sum_result = n1 + n2
```

```python
sub_result = n1 - n2
```

```python
print("First Number:", n1)
```

```python
print("Second Number:", n2)
```

```python
print("Addition:", sum_result)
```

```python
print("Subtraction:", sub_result)
```

```python
print("n1 == n2 :", n1 == n2)
```

```python
print("n1 > n2  :", n1 > n2)
```

```python
#18.	Write a Python program to demonstrate object composition, access composed objects, and display their details.
```

```python
class Address:
    def __init__(self, city, state):
        self.city = city
        self.state = state

    def display_address(self):
        print("City :", self.city)
        print("State:", self.state)
```

```python
class Employee:
    def __init__(self, emp_id, name, address):
        self.emp_id = emp_id
        self.name = name
        self.address = address  # Composition

    def display_employee(self):
        print("Employee ID:", self.emp_id)
        print("Employee Name:", self.name)
        print("Address Details:")
        self.address.display_address()
```

```python
addr = Address("Pune", "Maharashtra")
```

```python
emp = Employee(101, "Aditi", addr)
```

```python
emp.display_employee()
```

```python
#21.	Write a Python program to use the itertools module, generate permutations, combinations, and Cartesian products.
```

```python
from itertools import permutations, combinations, product
```

```python
items = ['A', 'B', 'C']
```

```python
print("Permutations:")
```

```python
for p in permutations(items, 2):
    print(p)
```

```python
print("\nCombinations:")
```

```python
for c in combinations(items, 2):
    print(c)
```

```python
print("\nCartesian Product:")
```

```python
for cp in product(items, repeat=2):
    print(cp)
```
